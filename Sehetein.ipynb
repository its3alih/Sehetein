```python
# برنامج بسيط لمحاكاة القائمة الهاتفية الصحية
# يجيب المواطن على أسئلة باختيار الأرقام

def health_check():
    print("مرحباً بك في نظام صحة قريب (Sehetein)")
    print("الرجاء الإجابة على بعض الأسئلة الصحية")

    # السؤال الأول: العمر
    age = int(input("كم عمرك؟ أدخل رقم: "))

    # السؤال الثاني: الأعراض
    print("\nهل لديك واحد من هذه الأعراض؟")
    print("1. حمى")
    print("2. سعال")
    print("3. ضيق تنفس")
    print("4. لا يوجد أي من هذه")
    symptom = int(input("أدخل الرقم المناسب: "))

    # السؤال الثالث: علامات خطورة
    print("\nهل لديك أي من علامات الخطر؟")
    print("1. نزيف شديد")
    print("2. فقدان وعي أو إغماء")
    print("3. ألم شديد في الصدر")
    print("4. لا يوجد")
    danger = int(input("أدخل الرقم المناسب: "))

    # المنطق البسيط للتقييم
    if danger in [1, 2, 3]:
        result = "أحمر"
        advice = "حالتك خطيرة. الرجاء التوجه فوراً إلى أقرب مستشفى."
    elif symptom in [1, 2, 3]:
        result = "برتقالي"
        advice = "أنت بحاجة لمراجعة طبيب أو الاتصال بالعامل الصحي في القرية."
    else:
        result = "أخضر"
        advice = "لا توجد علامات خطيرة. حافظ على الراحة وتابع الأعراض."

    print("\n-------------------------")
    print(f"تصنيف حالتك: {result}")
    print(f"النصيحة الطبية: {advice}")
    print("-------------------------")

# تشغيل البرنامج
health_check()
```

```python
import random

# برنامج محاكاة القائمة الهاتفية الصحية (SehaNear)
# باللغة العربية - نسخة موسعة

def health_check():
    print("مرحباً بك في نظام صحة قريب (SehaNear)")
    print("الرجاء الإجابة على بعض الأسئلة باختيار الرقم المناسب\n")

    # السؤال 1: العمر
    age = int(input("كم عمرك؟ أدخل الرقم: "))

    # السؤال 2: الجنس
    print("\nاختر الجنس:")
    print("1. ذكر")
    print("2. أنثى")
    gender = int(input("أدخل الرقم: "))

    # السؤال 3: الحمى
    print("\nهل لديك حمى؟")
    print("1. نعم")
    print("2. لا")
    fever = int(input("أدخل الرقم: "))

    # السؤال 4: السعال
    print("\nهل لديك سعال؟")
    print("1. نعم")
    print("2. لا")
    cough = int(input("أدخل الرقم: "))

    # السؤال 5: ضيق التنفس
    print("\nهل تشعر بضيق في التنفس؟")
    print("1. نعم")
    print("2. لا")
    breath = int(input("أدخل الرقم: "))

    # السؤال 6: آلام في الصدر
    print("\nهل لديك ألم شديد في الصدر؟")
    print("1. نعم")
    print("2. لا")
    chest_pain = int(input("أدخل الرقم: "))

    # السؤال 7: نزيف
    print("\nهل لديك نزيف شديد؟")
    print("1. نعم")
    print("2. لا")
    bleeding = int(input("أدخل الرقم: "))

    # السؤال 8: فقدان وعي
    print("\nهل فقدت وعيك أو أغمي عليك مؤخراً؟")
    print("1. نعم")
    print("2. لا")
    unconscious = int(input("أدخل الرقم: "))

    # السؤال 9: الحمل (للسيدات فقط)
    pregnant = 0
    if gender == 2 and 15 <= age <= 50:
        print("\nهل أنتِ حامل؟")
        print("1. نعم")
        print("2. لا")
        pregnant = int(input("أدخل الرقم: "))

    # السؤال 10: الأمراض المزمنة
    print("\nهل لديك أمراض مزمنة (مثل الضغط، السكر، القلب)؟")
    print("1. نعم")
    print("2. لا")
    chronic = int(input("أدخل الرقم: "))

    # -------------------------------
    # المنطق البسيط للتقييم
    danger_signs = [chest_pain, bleeding, unconscious]
    mild_symptoms = [fever, cough, breath]

    if 1 in danger_signs:
        result = "أحمر"
        advice = "حالتك خطيرة. يجب التوجه فوراً إلى أقرب مستشفى أو الاتصال بالإسعاف."
    elif 1 in mild_symptoms or chronic == 1 or pregnant == 1:
        result = "برتقالي"
        advice = "ننصحك بمراجعة أقرب عيادة أو التواصل مع العامل الصحي في القرية خلال 24 ساعة."
    else:
        result = "أخضر"
        advice = "لا توجد علامات خطيرة حالياً. يُرجى الراحة، شرب سوائل كافية، ومتابعة الأعراض."

    # إنشاء كود إحالة Referral Code (6 أرقام عشوائية)
    referral_code = random.randint(100000, 999999)

    # -------------------------------
    # إخراج النتيجة
    print("\n-------------------------")
    print(f"تصنيف حالتك: {result}")
    print(f"النصيحة الطبية: {advice}")
    print(f"كود الإحالة الخاص بك: {referral_code}")
    print("يمكنك إعطاء هذا الكود للعامل الصحي أو الطبيب لمتابعة حالتك.")
    print("-------------------------")

# تشغيل البرنامج
health_check()
h
```
